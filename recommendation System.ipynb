```python
import pandas as pd
movie_data=pd.read_csv("movies.csv")
movie_data.head()
```

```python

movie_data['year']=movie_data.title.str.extract('(\(\d\d\d\d\))',expand=False)
movie_data['year']=movie_data.year.str.extract('(\d\d\d\d)',expand=False)
movie_data['title']=movie_data.title.str.replace("(\(\d\d\d\d\))",' ')
movie_data['title']=movie_data['title'].apply(lambda x:x.strip())
movie_data.head()
```

```python
movie_data['genres']=movie_data['genres'].apply(lambda x: x.split('|'))
movie_data.head()
```

```python
movie_data_pro=movie_data.copy()

for index, row in movie_data_pro.iterrows():
    for genre in row["genres"]:
        movie_data_pro.at[index,genre]=1
        
movie_data_pro=movie_data_pro.fillna(0)
movie_data_pro.head()
```

```python
rating_data=pd.read_csv("ratings.csv")
rating_data.head()
```

```python
rating_data.drop(['timestamp'],axis=1,inplace=True)
rating_data.head()
```

```python
input_user=pd.DataFrame([['Breakfast Club, The',5],['Toy Story',3.5],['Jumanji',2],['Pulp Fiction',5],['Akira',4.5]])
input_user.columns=["title","rating"]
input_user
```

```python
copy_movie_pro=movie_data_pro.copy()
copy_movie_pro.set_index('title',inplace=True)
codedmatrix=copy_movie_pro.loc[input_user['title'],]
codedmatrix.drop(['movieId','genres','year'],axis=1,inplace=True)
codedmatrix.index.name=None
codedmatrix.reset_index(inplace=True)
codedmatrix.drop(['index'],axis=1,inplace=True)
codedmatrix

```

```python
weighted_matrix=codedmatrix.transpose().dot(input_user["rating"])
weighted_matrix
```

```python
user_profile=weighted_matrix/weighted_matrix.sum()
user_profile
```

```python
data_coded=movie_data_pro.copy()
data_coded.drop(['movieId','genres','year'],axis=1,inplace=True)
data_coded.set_index('title',inplace=True)
data_coded.index.name=None
data_coded=data_coded.transpose()
data_coded.reset_index(inplace=True)
data_coded.drop(['index'],axis=1,inplace=True)
user_pro=user_profile.copy()
user_pro=pd.DataFrame(user_pro)
user_pro.reset_index(inplace=True)
user_pro.drop(['index'],axis=1,inplace=True)

recommendation_matrix=data_coded.transpose().dot(user_pro)
recommendation_matrix.columns=["recommendations"]
recommendation_matrix=recommendation_matrix.sort_values('recommendations',ascending=False)
recommendation_matrix.head()
```
